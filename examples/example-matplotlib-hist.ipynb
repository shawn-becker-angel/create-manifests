```python
## EXAMPLE seaborn distplot

# from https://www.codespeedy.com/histogram-of-an-image-using-matplotlib-in-python/

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

# read and show a given image:
image_file = 'angel-studios.png'

#image read function
img=mpimg.imread(image_file)
#image sclicing into 2D. 
x=img[:,:,0]
# x co-ordinate denotation. 
plt.xlabel("Value")
# y co-ordinate denotation.
plt.ylabel("pixels Frequency")
# title of an image .
plt.title("Original Image")
# imshow function with comperision of gray level value.
plt.imshow(x,cmap="gray")
#plot the image on a plane.
plt.show()

import seaborn as sns
sns.distplot(x, bins=50)


# plt.title("HIstogramm for given Image'  ")
# plt.xlabel("Value")
# plt.ylabel("pixels Frequency")
# #hist function is used to plot the histogram of an image.
# plt.hist(x)
```
